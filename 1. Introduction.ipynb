```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 1, 0.01)
y = np.sin(2 * np.pi * x)
noise = np.random.normal(0,0.3,10)
```

```python
sample_x = np.linspace(0, 1, num=10)
```

```python
sample_y = np.sin(2 * np.pi * sample_x) + noise
```

```python
plt.plot(x, y, c='lime')
plt.scatter(sample_x, sample_y, s=90, facecolors='none', lw=2,edgecolors='b')
plt.xlabel('x')
plt.ylabel('t')
plt.show()
```
